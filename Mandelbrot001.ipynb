```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the Mandelbrot set
xmin, xmax, ymin, ymax = -2, 1, -1.5, 1.5
maxiter = 100
xres, yres = 500, 500

# Create an array of complex numbers representing the grid of points in the complex plane
x = np.linspace(xmin, xmax, xres)
y = np.linspace(ymin, ymax, yres)
c = x[:, np.newaxis] + 1j*y[np.newaxis, :]

# Initialize the array of iteration counts
z = c.copy()
iter_counts = np.zeros_like(c, dtype=int)

# Compute the Mandelbrot set by iterating the function z = z**2 + c for a maximum of maxiter iterations
for i in range(maxiter):
    z = z**2 + c
    diverged = np.abs(z) > 2
    iter_counts[diverged] = i

# Plot the Mandelbrot set with a grayscale colormap
plt.imshow(iter_counts.T, cmap='hsv', extent=(xmin, xmax, ymin, ymax), interpolation='bicubic')

# Set the x and y limits of the plot to match the range of the Mandelbrot set
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)

# Display the plot
plt.show()
```
