```python
import numpy as np
import matplotlib.pyplot as plt

# generate x-values
x = np.linspace(-5, 5, 500)

# define six different functions
y1 = np.sin(x)
y2 = x**3
y3 = np.exp(-x**2)
y4 = np.tanh(x)
y5 = np.abs(x)
y6 = np.sqrt(np.abs(x))

# create a 3x2 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(10, 5))

# plot each function on a different subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('y = sin(x)')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('y = x^3')
axs[0, 2].plot(x, y3)
axs[0, 2].set_title('y = exp(-x^2)')
axs[1, 0].plot(x, y4)
axs[1, 0].set_title('y = tanh(x)')
axs[1, 1].plot(x, y5)
axs[1, 1].set_title('y = |x|')
axs[1, 2].plot(x, y6)
axs[1, 2].set_title('y = sqrt(|x|)')

# set common x and y labels for all subplots
for ax in axs.flat:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(True)

# adjust spacing between subplots
fig.tight_layout()

# display the plot
plt.show()
```
