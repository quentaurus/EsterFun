```python
import numpy as np
import matplotlib.pyplot as plt

# generate data
x = np.linspace(0, 4*np.pi, 1000)   # x-axis values from 0 to 4pi
yt = (np.sin(x)**2 - np.cos(x)**2)
yn = np.cos(x)*np.sin(x)

# plot both sine and cosine waves
plt.plot(x, yt, 'r--', label='täljare')    # dashed red line for sine wave
plt.plot(x, yn, 'b:', label='nämnare')   # dotted blue line for cosine wave
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Waves')
plt.legend()
plt.show()
```
